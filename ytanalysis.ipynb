```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import isodate
```

**Importing Trending YouTube Videos Data, according to July-August 2024**

```python
df = pd.read_csv('trending_videos.csv')
df1 = pd.read_csv('category.csv')
```

```python
df
```

**Searching for null values and Clearing them**

```python
df.isnull().sum()
```

```python
df['description'].fillna('No Description', inplace=True)
```

```python
df1.columns = ['Id','Category']
df1
```

**Assigning Categories to their respective Category_id**

```python
df = pd.merge(df, df1, how='inner', left_on='category_id', right_on='Id')
df
```

**Highly Viewed Video in Current Trending List**

```python
view_max = df['view_count'].max()
df.loc[df['view_count']==view_max,['title','channel_title','view_count']]
```

**Highly Liked Video in Current Trending List**

```python
like_max = df['like_count'].max()
df.loc[df['like_count']==like_max,['title','channel_title','like_count']]
```

**Analyzing Trending Videos Dataset**

1. Classifying Trending Videos Category-wise 

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12, 8))
sns.countplot(y=df['Category'], order=df['Category'].value_counts().index)
plt.title('Number of Trending Videos by Category')
plt.xlabel('Number of Videos')
plt.ylabel('Category')
plt.savefig('ytAnalysis(1).png')
plt.show()
```


2. Relation between Publish Hours and Number of Videos Posted

```python
df['published_at'] = pd.to_datetime(df['published_at'])
```

```python
df['publish_hour'] = df['published_at'].dt.hour

plt.figure(figsize=(12, 6))
sns.countplot(x='publish_hour', data=df, palette='magma')
plt.title('Distribution of Videos by Publish Hour')
plt.xlabel('Publish Hour')
plt.ylabel('Number of Videos')
plt.savefig('ytAnalysis(2).png')
plt.show()
```


3. Effect of Publish Hours over View Counts

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='publish_hour', y='view_count', data=df, alpha=0.6, color='teal')
plt.title('Publish Hour vs View Count')
plt.xlabel('Publish Hour')
plt.ylabel('View Count')
plt.savefig('ytAnalysis(3).png')
plt.show()
```


4. Average Like count, View count and Comment count of each category

```python
categ = df.groupby('Category')[['view_count', 'like_count', 'comment_count']].mean().sort_values(by='view_count', ascending=False)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 10))

sns.barplot(y=categ.index, x=categ['view_count'].sort_values(ascending=False), ax=axes[0], palette='viridis')
axes[0].set_title('Average View Count by Category')
axes[0].set_xlabel('Average View Count')
axes[0].set_ylabel('Category')

sns.barplot(y=categ.index, x=categ['like_count'].sort_values(ascending=False), ax=axes[1], palette='viridis')
axes[1].set_title('Average Like Count by Category')
axes[1].set_xlabel('Average Like Count')

sns.barplot(y=categ.index, x=categ['comment_count'].sort_values(ascending=False), ax=axes[2], palette='viridis')
axes[2].set_title('Average Comment Count by Category')
axes[2].set_xlabel('Average Comment Count')

plt.tight_layout()
plt.savefig('ytAnalysis(4).png')
plt.show()
```

5. Relation between Video Length and View Counts

```python
df['duration'] = df['duration'].apply(lambda x: isodate.parse_duration(x).total_seconds())
```

```python
df['duration_range'] = pd.cut(df['duration'], bins=[0, 300, 600, 1200, 3600, 7200], labels=['0-5 min', '5-10 min', '10-20 min', '20-60 min', '60-120 min'])
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='duration', y='view_count', data=df, alpha=0.6, color='blue')
plt.title('Video Length vs View Count')
plt.xlabel('Video Length (in secs)')
plt.ylabel('View Count')
plt.savefig('ytAnalysis(5).png')
plt.show()
```

6. Relation between Average View Count, Like Count and Duration

```python
length_relation = df.groupby('duration_range')[['view_count', 'like_count']].mean()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18,8))

sns.barplot(y=length_relation.index, x=length_relation['view_count'], ax=axes[0], palette='magma')
axes[0].set_title('Average View Count by Duration Range')
axes[0].set_xlabel('Average View Count')
axes[0].set_ylabel('Duration Range')

# like count by duration range
sns.barplot(y=length_relation.index, x=length_relation['like_count'], ax=axes[1], palette='magma')
axes[1].set_title('Average Like Count by Duration Range')
axes[1].set_xlabel('Average Like Count')
axes[1].set_ylabel('')
plt.savefig('ytAnalysis(6).png')
```

7. Effects of Tags on View Count

```python
df['tag_count'] = df['tags'].apply(len)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='tag_count', y='view_count', data=df, alpha=0.6, color='cyan', edgecolor='black', linewidth=1)
plt.title('Number of Tags vs View Count')
plt.xlabel('Number of Tags')
plt.ylabel('View Count')
plt.savefig('ytAnalysis(7).png')
plt.show()
```
